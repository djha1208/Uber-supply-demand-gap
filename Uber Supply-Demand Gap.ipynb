```python
# Import the numpy,pandas,datetime,matplotlib.pyplot and seaborn packages
import numpy as np
import pandas as pd
import datetime as date
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Code for importing the csv file here
Data =pd.read_csv(r'C:\Users\user\Downloads\Uber Request Data.csv') 
Data
```

```python
# Code for inspection here
print(Data.columns)
print(Data.shape)
print(Data.describe())
print(type(Data.info()))
```

```python
# Convert Data type of Request timestamp to datetime
Data["Request timestamp"] = Data["Request timestamp"].astype(str)
Data['Request timestamp'] = Data['Request timestamp'].str.replace("/", "-")
Data['Request timestamp']=pd.to_datetime(Data['Request timestamp'], dayfirst=True)  # dayfirst is used to display day first in time format

# Convert Data type of Drop Request timestamp to datetime
Data["Drop timestamp"] = Data["Drop timestamp"].astype(str)
Data['Drop timestamp'] = Data['Drop timestamp'].str.replace("/", "-")
Data['Drop timestamp']=pd.to_datetime(Data['Drop timestamp'], dayfirst=True)       # dayfirst is used to display day first in time format
print(Data['Request timestamp'])
print(Data['Drop timestamp'])
```

```python
# Extract day and hour from Request timestamp
Data['Request Day']=Data['Request timestamp'].dt.day
Data['Request hour']=Data['Request timestamp'].dt.hour
print(Data['Request Day'])
print(Data['Request hour'])
Data
```

# Gap between Supply and Demand

## Overall Demand, Supply and Gap

```python
# Data1 is new dataframe
Data1=pd.DataFrame()
Data1=Data.copy()           # Copy 'Data' Dataframe in 'Data1' Dataframe
Data1['Status']='Demand'
Data2=Data.append(Data1)    # Append 'Data' and 'Data1'dataframe and store in 'Data2' dataframe
Data2['Status']=Data2['Status'].replace(['Cancelled','No Cars Available'],'Gap')
Data2['Status']=Data2['Status'].replace(['Trip Completed'],'Supply')

# Supply-Demand gap 
sns.countplot(x='Status',order=['Demand','Supply','Gap'],data=Data2,palette=sns.color_palette('Set2'))
plt.title('Supply-Demand gap')
plt.show()
```

## Calculation

```python
# Total Demand
Total_Demand=Data2[( Data2["Status"] == "Demand")] 
print(len(Total_Demand.index))

# Total Supply
Total_Supply=Data2[(Data2["Status"] == 'Supply')]
print(len(Total_Supply.index))

# Total Supply-Demand Gap
Total_Supply_Demand_Gap=Data2[(Data2["Status"] == 'Gap')]
print(len(Total_Supply_Demand_Gap.index))
```
